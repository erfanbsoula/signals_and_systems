```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
alpha = 0.5
x = [1, 1, 1, 1, 1]
ys = np.zeros((5, 20))
for i in range(0, 5):
    for n in range(i,20):
        ys[i][n] = x[i] * alpha**(n-i)

n_range = range(0, 20)
```

```python
plt.stem(n_range, ys[0])
plt.title("(0.5^(n)) * u[n]")
plt.show()
```

```python
plt.stem(n_range, ys[1])
plt.title("(0.5^(n-1)) * u[n-1]")
plt.show()
```

```python
plt.stem(n_range, ys[2])
plt.title("(0.5^(n-2)) * u[n-2]")
plt.show()
```

```python
plt.stem(n_range, ys[3])
plt.title("(0.5^(n-3)) * u[n-3]")
plt.show()
```

```python
plt.stem(n_range, ys[4])
plt.title("(0.5^(n-4)) * u[n-4]")
plt.show()
```

```python
y = ys[0] + ys[1] + ys[2] + ys[3] + ys[4]
plt.stem(n_range, y)
plt.title("final response")
plt.show()
```
