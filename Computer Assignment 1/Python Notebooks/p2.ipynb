```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import pi
```

```python
def delta_function(n):
    if n == 0:
        return 1;
    else:
        return 0;

def step_function(n):
    if n >= 0:
        return 1;
    else:
        return 0;
```

```python
def system_a(x, n):
    return 0.5*(x(n)+x(n-1))

def system_b(x, n):
    return x(n+1) - x(n)

def system_c(x, n):
    return x(n+1) - 2*x(n) + x(n-1)

def system_d(x, n):
    res = 0;
    for m in range(0, 20):
        res += np.cos((m*pi)/10)*x(n-m)

    return res
```

```python
n_start = -5
n_stop = 30
step_length = 1
n_range = np.arange(n_start, n_stop, step_length)
count = n_range.size

H_a = np.zeros(count)
H_b = np.zeros(count)
H_c = np.zeros(count)
H_d = np.zeros(count)

for n in n_range:
    H_a[n-n_start] = system_a(delta_function, n)
    H_b[n-n_start] = system_b(delta_function, n)
    H_c[n-n_start] = system_c(delta_function, n)
    H_d[n-n_start] = system_d(delta_function, n)
```

```python
plt.stem(n_range, H_a)
plt.title("System A - Impulse Response")
plt.show()
```

```python
plt.stem(n_range, H_b)
plt.title("System B - Impulse Response")
plt.show()
```

```python
plt.stem(n_range, H_c)
plt.title("System C - Impulse Response")
plt.show()
```

```python
plt.stem(n_range, H_d)
plt.title("System D - Impulse Response")
plt.show()
```
