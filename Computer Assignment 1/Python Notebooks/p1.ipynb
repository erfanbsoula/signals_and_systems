```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import pi #to have it all along the program
```

# Part a

```python
k = 0.125
T = 1/k
# sampling frequency:
sf = 1
n = np.arange(0, 50, 1/sf)
start = int(T * sf)
stop = n.size - start
indxs = range(start, stop)
h_axis = n[start:stop] - T
```

```python
x = np.cos(2*k*pi*n)
plt.stem(h_axis, x[start:stop])
plt.xlabel("n")
plt.title("x[n]")
#plt.savefig("p1_x1.jpg", dpi=360)
plt.show()
```

## A.

```python
y = [(1/T)*(x[i+1]-x[i]) for i in indxs]
plt.stem(h_axis, y)
plt.xlabel("n")
plt.title("(1/T)*( x[n+1]-x[n] )")
plt.show()
```

## B.

```python
y = [(1/(2*T))*(x[i+1]-x[i-1]) for i in indxs]
plt.stem(h_axis, y)
plt.xlabel("n")
plt.title("( 1/(2*T) )*( x[n+1]-x[n-1] )")
plt.show()
```

## C.

```python
y = [(1/(T**2))*(x[i+1]-2*x[i]+x[i-1]) for i in indxs]
plt.stem(h_axis, y)
plt.xlabel("n")
plt.title("( 1/(T^2) )*( x[n+1] - 2*x[n] + x[n-1] )")
plt.show()
```

## D.

```python
y = [(0.1*x[i-2]+0.2*x[i-1]+0.4*x[i]+0.2*x[i+1]+0.1*x[i+2]) for i in indxs]
plt.stem(h_axis, y)
plt.xlabel("n")
plt.title("0.1*x[n-2]+0.2*x[n-1]+0.4*x[n]+0.2*x[n+1]+0.1*x[n+2]", fontsize=11)
plt.show()
```
